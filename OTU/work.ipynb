```python
import collections
import matplotlib.pyplot
%matplotlib inline
```

```python
file_path_list = [
    'OTU/files/18599.ktg',
    'OTU/files/18600.ktg',
    'OTU/files/18601.ktg',
    'OTU/files/18602.ktg',
    'OTU/files/18603.ktg',
]
file_list = list()
for f in file_path_list:
    with open(f, 'r') as fi:
        body = fi.readlines()
        a = str()
        for line in body:
            a += line
        file_list.append(a)
```

```python
text_len_list = list()
for text in file_list:
    print(len(text), list(set(text)))
    text_len_list.append(len(text))
```

```python
def sort_second(val):    
    return val[1]


def count_gram(text, len_gram):
    gram_list = list()
    index_1 = 0
    index_2 = len_gram
    while True:
        temp = text[index_1:index_2]
        index_1 += len_gram
        index_2 += len_gram
        if temp == '':
            break
        else:
            gram_list.append(temp)
            
    gram_count = collections.Counter(gram_list)
    
    gram_count_list = list()
    for i in gram_count.keys():
        gram_count_list.append((i, gram_count[i]))
    gram_count_list.sort(key=sort_second, reverse=True)
    return gram_count_list
```

```python
count_gram(file_list[0], 2)
```

```python
def plotter(gram_count, stopper=9, name=None):
    s = list()
    names = list()
    ax = matplotlib.pyplot.gca()
    for index, i in enumerate(gram_count):
        s.append(i[0])
        names.append(i[1])
        if index > stopper:
            break
    if name is not None:
        ax.set_title(name)
    # x = range(len(s))
    ax.bar(s, names, align='edge')
    # ax.set_xticks(x)
    ax.set_xticklabels(s)
    matplotlib.pyplot.show()
```

```python
top_len = 10
print('Top-{0}'.format(top_len))
for text_number in range(len(file_list)):
    for len_gram in [2, 3]:
        temp_gram = count_gram(file_list[text_number], len_gram)
        temp_name = 'Text {0}, {1}-gramm. Text len = {2}'.format(
            text_number, len_gram, text_len_list[text_number]
        )
        plotter(temp_gram, name=temp_name, stopper=top_len-1)
```
